```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample latency data reflecting observations during testing
data = {
    'Load Level': ['Low', 'Medium', 'High', 'Extreme'],
    'Average Latency (ms)': [100, 150, 300, 450],  
    'Spikes Observed (ms)': [120, 160, 350, 500]   
}

# Create a DataFrame
latency_df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(12, 6))

# Bar chart for average latency
plt.bar(latency_df['Load Level'], latency_df['Average Latency (ms)'], color='skyblue', label='Average Latency')

# Line plot for observed spikes
plt.plot(latency_df['Load Level'], latency_df['Spikes Observed (ms)'], color='red', marker='o', label='Spikes Observed')

# Adding titles and labels
plt.title('Latency Comparison Under Different Load Levels', fontsize=14)
plt.xlabel('Load Level', fontsize=12)
plt.ylabel('Latency (ms)', fontsize=12)
plt.ylim(0, 600)  # Set y-axis limits for better visibility
plt.grid(axis='y')

# Show the values on top of the bars
for index, value in enumerate(latency_df['Average Latency (ms)']):
    plt.text(index, value + 10, str(value), ha='center', va='bottom', fontsize=10)

# Show the values for observed spikes
for index, value in enumerate(latency_df['Spikes Observed (ms)']):
    plt.text(index, value + 10, str(value), ha='center', va='bottom', fontsize=10, color='red')

# Adding legend
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
