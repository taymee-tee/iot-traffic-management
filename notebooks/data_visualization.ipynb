# Label Distribution Visualization Script

## Overview
This script visualizes the distribution of labels (0 for Normal and 1 for attack) in the processed datasets for IoT devices (Archer, Camera, and Indoor).

## Libraries Used
- `pandas`: For data manipulation and loading datasets.
- `matplotlib.pyplot`: For creating plots.
- `seaborn`: For enhanced visualization aesthetics.

## Functionality
### `visualize_label_distribution(device_name)`
This function performs the following tasks:
- Loads the processed dataset for the specified IoT device.
- Creates a count plot to visualize the distribution of labels (0 and 1).
- Customizes the plot with appropriate labels and a title.

## Execution
The script visualizes the label distribution for the following devices:
- Archer
- Camera
- Indoor


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define file paths for the processed datasets
file_paths = {
    'archer': r'C:\Users\USER\IoT_Network_Traffic_Management\data\processed_data\processed_archer.csv',
    'camera': r'C:\Users\USER\IoT_Network_Traffic_Management\data\processed_data\processed_camera.csv',
    'indoor': r'C:\Users\USER\IoT_Network_Traffic_Management\data\processed_data\processed_indoor.csv'
}

# Function to visualize label distribution
def visualize_label_distribution(device_name):
    """
    Load the processed dataset for a specific device and visualize the distribution of labels (0 and 1).

    Parameters:
    device_name (str): The name of the IoT device (e.g., 'archer', 'camera', 'indoor').
    """
    
    # Load the processed dataset
    df = pd.read_csv(file_paths[device_name])
    
    # Set the style of seaborn
    sns.set(style='whitegrid')

    # Create a count plot for the label distribution
    plt.figure(figsize=(8, 5))
    sns.countplot(x='label', data=df, palette='viridis')

    # Add labels and title
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.title(f'Label Distribution for {device_name.capitalize()} Device')
    plt.xticks(ticks=[0, 1], labels=['Normal (0)', 'Attack (1)'])

    # Show the plot
    plt.show()

# Visualize label distribution for Archer, Camera, and Indoor
for device in ['archer', 'camera', 'indoor']:
    visualize_label_distribution(device)
```
